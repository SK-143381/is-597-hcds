```python
BEST WORKING VERSION

api_key = 'c0826860be16ee211a9e15b22d059034'

# Global flag to control the beep loop
beep_playing = threading.Event()

def play_beep():
    """Plays beep sound repeatedly until stopped."""
    while not beep_playing.is_set():
        os.system(f'afplay "beeping.mp3"')  # Use 'start beep.mp3' on Windows or 'mpg123 beep.mp3' on Linux
        time.sleep(1)  # Adjust as necessary for the length of your beep sound

def speak(text):
    """Converts text to speech and plays it."""
    if text.strip():  # Check if the text is not empty or whitespace
        text_for_speaking = re.sub(r'[.!?]', '', text)
        tts = gTTS(text=text_for_speaking, lang='en')
        tts.save('output.mp3')
        os.system(f'afplay output.mp3')  # Adjust command based on your OS
    else:
        print("No text to speak.")

def listen():
    """Listens for speech and converts it to text."""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        speak("Listening...")  # Play 'Listening...' prompt before capturing audio
        print("\nListening...")
        audio = r.listen(source)
    try:
        user_input = r.recognize_google(audio)
        # This line prints what the speech recognition library recognized.
        print(f"The system heard: '{user_input}'")
        speak(f"The system heard: '{user_input}'")  # Speak back what was heard
        return user_input
    except sr.UnknownValueError:
        print("Sorry, I could not understand your speech.")
        return ""
    except sr.RequestError:
        print("Sorry, I am unable to access the Google Speech Recognition service.")
        return ""

def read_accessibility_laws_csv(filename):
    """Reads a CSV file into a list of dictionaries for accessibility laws."""
    with open(filename, mode='r', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=';')
        return [row for row in reader]

def clean_punctuation_spaces(text):
    """Removes unnecessary spaces before punctuation marks like apostrophes."""
    return re.sub(r'\s([?.!,\'])', r'\1', text)

def chat_with_ollama(accessibility_laws_data):
    """Manages the chat conversation with Ollama."""
    print("Starting chat with Ollama. Type 'quit' to exit.")
    messages = []

    # Ask the admin for the system response
    system_prompt = input("Admin, enter the initial system prompt for Ollama: ")
    system_prompt += " You are a social companion that provides advice in three domains: physical activity, mental health, and workplace accommodations. You must refer to the accessibility laws data to find accurate information, and incorporate it into your responses. Use the summary provided and mention years of the laws. If they are being discriminated against, or have any other problem, please ask them what the problem is first, before giving any advice. Be very concise with your response, since you want to help not ramble."

    if system_prompt.strip():
        # Ensure that system prompts are non-empty before appending
        messages.append({'role': 'system', 'content': system_prompt})

        if accessibility_laws_data:
            # Append accessibility laws data to the system response
            laws_content = '\n'.join(f"{law['Name']} ({law['Year']}): {law['Summary']}" for law in accessibility_laws_data if 'Name' in law and law['Summary'].strip())
            if laws_content.strip():
                messages.append({'role': 'system', 'content': laws_content})
    
    while True:
        # Listen to user input
        user_input = listen()
        if user_input.lower() == 'quit':
            break

        if user_input.strip():
            messages.append({'role': 'user', 'content': user_input})

            # Start playing beep sound in a loop on a separate thread
            beep_playing.clear()
            beep_thread = threading.Thread(target=play_beep)
            beep_thread.start()

            try:
                # Generate response from Ollama based on the user input
                stream = ollama.chat(model='llama3.2', messages=messages, stream=True)
                complete_response = ""  # Accumulate the complete response for speaking
                for chunk in stream:
                    response = chunk['message']['content']
                    complete_response += " " + response

                # Clean up spaces before punctuation
                complete_response = clean_punctuation_spaces(complete_response)

                # Print the response word by word
                words = complete_response.split()
                for word in words:
                    print(word, end=' ', flush=True)
                    time.sleep(0.2)

            except Exception as e:
                print(f"Error during Ollama chat: {str(e)}")
                break

            # Stop the beep sound and wait for the thread to finish
            beep_playing.set()
            beep_thread.join()

            # Speak the complete response immediately after generating it
            if complete_response.strip():
                speak(complete_response.strip())  # Play Ollama's response

            # Start "Listening..." again after the response is spoken (this line was commented out in the last version)
        else:
            print("Message is empty, please enter some text.")

if __name__ == "__main__":
    # Reading file names from system input
    accessibility_laws_file = 'accessibility-laws.csv'
    
    # Reading the contents of the accessibility laws CSV
    accessibility_laws_data = read_accessibility_laws_csv(accessibility_laws_file)
    
    # Now you can use accessibility_laws_data in your application
    chat_with_ollama(accessibility_laws_data)
```
